# Importación de librerias

```python
import pandas as pd
import plotly.offline as pyo
import plotly.graph_objs as go
```

# Scatter plot

```python
df_iris = pd.read_csv('datasets/iris.csv')
df_iris.head(3)
```

```python
data1 = [go.Scatter(x=df_iris['sepal.length'], 
                    y = df_iris['sepal.width'],
                    mode = "markers")]

layout1 = go.Layout(title="Iris scatter plot",
                     xaxis={"title":"sepal length"},
                     yaxis={"title":"sepal width"})
```

```python
fig = go.Figure(data=data1, layout=layout1)
pyo.plot(fig, filename="plot_iris_scatter.html")
```

# Lines plot

```python
df_sp500 = pd.read_csv('datasets/SP500_data.csv')
df_sp500.head(3)
```

```python
traza1 = go.Scatter(x = df_sp500['Date'], 
                    y = df_sp500['Open'],
                    mode = "lines+markers")

traza2 = go.Scatter(x = df_sp500['Date'],
                    y = df_sp500['Close'],
                    mode = 'lines+markers')
data2 = [traza1, traza2]

layout2 = go.Layout(title = 'SP500 line plot',
                    xaxis = dict(title = 'date'),
                    yaxis = dict(title = 'close'))
```

```python
fig = go.Figure(data = data2, layout=layout2)
pyo.plot(fig, filename = "plot_sp500_line_plot.html")
```

# Bar plot

```python
df_ventas = pd.read_csv('datasets/Ventas.csv', sep = ';')
df_ventas.head(10)
```

```python
traza1 = go.Scatter(x = df_ventas['Fecha'],
                    y = df_ventas['Total Venta Classic Cars'],
                    mode = 'lines',
                    name = 'Classic Cars')
traza2 = go.Scatter(x = df_ventas['Fecha'],
                    y = df_ventas['Total Venta Motorcycles'],
                    mode = 'lines',
                    name = 'Motorcycles')
data3 = [traza1, traza2]

layoud3 = go.Layout(title = 'Line plot Ventas Classic Cars vs Motorcycles',
                    xaxis = dict(title = 'Fecha'),
                    yaxis = dict(title = 'Total ventas'))
fig = go.Figure(data = data3, layout=layoud3)
```

```python
pyo.plot(fig, filename='lines_plot_Ventas_Classic_cars_vs_Motorcycles.html')
```

```python
df_medallas = pd.read_excel('datasets/Medallas_olimpicas.xlsx')
df_medallas.head(10)
```

```python
df_medallas = df_medallas.sort_values(by = 'Total', ascending = False)
```

```python
data4 = [go.Bar(x = df_medallas['País'], 
                y = df_medallas['Total'])]
layout4 = go.Layout(title = 'Medallas por país',
                    xaxis = dict(title = "País"),
                    yaxis = dict(title = 'Total medallas'))
fig = go.Figure(data = data4, layout=layout4)
```

```python
pyo.plot(fig, filename='plot_bar_medallas.html')
```

```python
traza1 = go.Bar(x = df_medallas['País'],
                y = df_medallas['Oro'],
                marker = {"color":"#FFD700"},
                name = 'Oro')
traza2 = go.Bar(x = df_medallas['País'],
                y = df_medallas['Plata'],
                marker = {"color":"#c0c0c0"},
                name = 'Plata')
traza3 = go.Bar(x = df_medallas['País'],
                y = df_medallas['Bronce'],
                marker = {"color":"#CD7F32"},
                name = 'Bronce')

data5 = [traza1, traza2, traza3]

layout5 = go.Layout(title = 'Medallas por país bar plot',
                    xaxis = dict(title = 'País'),
                    yaxis = dict(title = 'Medallas'),
                    barmode = 'stack')

fig = go.Figure(data=data5, layout=layout5)
```

```python
pyo.plot(fig, filename='plot_bar_medallas_trazas.html')
```

# Bubble plot

```python
df_info_pais = pd.read_excel('datasets/Info_pais.xlsx')
df_info_pais.head(5)
```

```python
df_info_pais['Continente'].unique()
```

```python
dic_cont_color = {'North America':'green', 'South America': 'green', 'Oceania': 'brown', 'Europe':'blue', 'Africa': 'gray', 'Asia': 'red'}
```

```python
df_info_pais['color'] = df_info_pais['Continente'].map(dic_cont_color)
df_info_pais[:10]
```

```python
data6 = [go.Scatter(x = df_info_pais['Renta per capita'], 
                    y = df_info_pais['Esperanza de vida'],
                    text = df_info_pais['País'],
                    mode = 'markers',
                    marker = dict(
                        size = df_info_pais['Poblacion']/5000000,
                        color = df_info_pais['color']
                    ))]

layout6 = go.Layout(title='Bubble plot Renta per capita vs Esperanza de vida',
                    xaxis = dict(title = 'Renta per capita'),
                    yaxis = dict(title = 'Esperanza de vida'))

fig = go.Figure(data = data6, layout=layout6)
```

```python
pyo.plot(fig, filename = 'bubble_plot.html')
```

# Histogram

```python
df_temperaturas2 = pd.read_excel('datasets/Temperaturas.xlsx')
df_temperaturas2.head(5)
```

```python
data9 = [go.Histogram(x = df_temperaturas2['T_Promedio'], 
                      xbins=dict(start = 0, end = 40, size = 5), 
                      histnorm='probability',
                      cumulative_enabled = True)]

layout9 = go.Layout(title = 'Histogram T_Promedio')

fig = go.Figure(data=data9, layout=layout9)
```

```python
pyo.plot(fig, filename='Histogram_plot.html')
```
